# Proses Scraping

```python
def getListPerusahaan(no_perusahaan,halaman):
    URL = "https://www.qerja.com/perusahaan?keyword=&industry_id="+str(no_perusahaan)+"&page="+str(halaman)+""
    
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    tables = soup.find_all("table")
    table = tables[0]
    
    datalist = []
    for k in table.find_all("div",{"class":"media-body"}):
        link = k.find_all("a",href=True)[0].get("href")
        name = k.find_all("a",href=True)[0].text.strip()
        datalist.append([name,link])
        
    df = pd.DataFrame(datalist)
    df.columns = ['Companyname','Link']
    
    return df
```

```python
max_halaman = 466
no_perusahaan = 19

tab_data = []

for k in range(max_halaman+1):
    tab_data.append(getListPerusahaan(no_perusahaan,k))
```

```python
df_BahanDanKimiaIndustri = pd.concat(tab_data)
df_BahanDanKimiaIndustri.reset_index(inplace=True,drop=True)
```

# Grab All Company

```python
def getDBCompany(halaman):
    URL = "https://www.qerja.com/perusahaan?keyword=&industry_id=&page="+str(halaman)+""
    
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    tables = soup.find_all("table")
    table = tables[0]
    
    datalist = []
    for k in table.find_all("div",{"class":"media-body"}):
        link = k.find_all("a",href=True)[0].get("href")
        name = k.find_all("a",href=True)[0].text.strip()
        datalist.append([name,link])
        
    df = pd.DataFrame(datalist)
    df.columns = ['Companyname','Link']
    
    return df
```

```python
max_halaman = 2661

tab_data = []

for k in range(max_halaman+1):
    tab_data.append(getDBCompany(k))
```

```python
df_all = pd.concat(tab_data)
df_all.reset_index(inplace=True,drop=True)
```

```python
df_all
```

# Company Type

```python
def getCompanyType(link_company):
    page = requests.get(link_company)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    result = []
    for k in soup.find_all("div",{"class":"col-xs-12"}):
        res = k.find_all("dl",{"class":"dl-horizontal"})
        if len(res) > 0:
            title = [k.text for k in res[0].find_all("dt")]
            value = [k.text for k in res[0].find_all("dd")]
            df = pd.DataFrame(value).T
            df.columns = title
            result.append(df)
    return df
```

```python
ls_type_comp = []

for k in df_all['Link'].values.tolist():
    ls_type_comp.append(getCompanyType(k))
```

```python
df_type_comp = pd.concat(ls_type_comp)
df_type_comp.reset_index(inplace=True,drop=True)
```

```python
new_columns_name = [k.strip().replace(":",'').strip() for k in list(df_type_comp.columns)]
df_type_comp.columns = new_columns_name
```
